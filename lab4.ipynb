```
import string

fn = open('/content/215-0.txt')
white=string.whitespace
punc=string.punctuation
speci=white+punc


def remove_pun(wrd):
  removed_list = list()
  total_removed_list = list()
  total_wrd =  wrd
  for n in total_wrd:
    for wrd_sp in n:
      wrd_sp1 = wrd_sp.strip(speci)
      removed_list.append(wrd_sp1)
    total_removed_list.append(removed_list)
  return total_removed_list

def total_list(wrd):
  total_wrd = list()
  for n in fn:
    char_list = list(n.split())
    total_wrd.append(char_list)
  return total_wrd


lst1 = remove_pun(total_list(fn))[0]
total_no_words = len(lst1)


```

```
#Commom and Unique words function
def com_uni_words(lst):
  word_dict = dict()
  word_lst = list()
  for i in lst:
    if i not in word_lst:
      word_dict[i] = 1
      word_lst.append(i)
    else:
      word_dict[i] = word_dict[i] + 1
  return word_dict


word_count_dict = com_uni_words(lst1)
print(word_count_dict)
```

```
#Different words used in dic

def diff_words(dic):
  diff_words_lst = list()
  for k in dic:
    diff_words_lst.append(k)
  return diff_words_lst


print('different words in book are',len(word_count_dict))
diff_words_lst_in_book = diff_words(word_count_dict)
```

```
#list of words into list:
def word_file_list():
  wrd = open('/content/words.txt')
  word_file_words = list()
  for i in wrd:
    wrds = i.strip()
    word_file_words.append(wrds)

  return word_file_words

word_file_words = set(word_file_list())

#function to compare words in the dictnary with words file

def comp_book_words(dic):
  wrd_not_in_words_file = list()
  wrd_in_words_file = list()
  for i in dic:
    if i not in word_file_words:
      wrd_not_in_words_file.append(i)
    else:
      wrd_in_words_file.append(i)
  return wrd_not_in_words_file


wrd_not_in_words_file = comp_book_words(diff_words_lst_in_book)
print(wrd_not_in_words_file)
```
